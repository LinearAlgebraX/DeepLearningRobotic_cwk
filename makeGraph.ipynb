```python
import matplotlib.pyplot as plt
from scipy import interpolate
import numpy as np

hyperP1 = [0.0001, 0.001, 0.01]
hyperP2 = [20, 60, 150]
hyperP3 = [0.5, 0.8, 0.99]
hyperP4 = [5e-6, 1e-5, 5e-4]


f1 = open("logs/lr.txt", "r", encoding="utf-8")
f2 = open("logs/bs.txt", "r", encoding="utf-8")
f3 = open("logs/mo.txt", "r", encoding="utf-8")
f4 = open("logs/we.txt", "r", encoding="utf-8")

F1 = f1.readlines()
F2 = f2.readlines()
F3 = f3.readlines()
F4 = f4.readlines()

f1.close()
f2.close()
f3.close()
f4.close()

F1 = [line.strip() for line in F1 if line.strip()!=""]
F1 = [line.split(" ") for line in F1]
F2 = [line.strip() for line in F2 if line.strip()!=""]
F2 = [line.split(" ") for line in F2]
F3 = [line.strip() for line in F3 if line.strip()!=""]
F3 = [line.split(" ") for line in F3]
F4 = [line.strip() for line in F4 if line.strip()!=""]
F4 = [line.split(" ") for line in F4]


plt.figure(figsize=(20, 13))
plt.figure(1)

ax1 = plt.subplot(221)
ax1.set_title("lr")
ax1.plot(hyperP1, list(map(float,[F1[0][0],F1[1][0],F1[2][0]])), label="accuracy")
ax1.plot(hyperP1, list(map(float,[F1[0][1],F1[1][1],F1[2][1]])), label="loss")
# plt.xticks(Cx[0])
plt.xlabel("lr")
plt.legend()

ax2 = plt.subplot(222)
ax2.set_title("bs")
ax2.plot(hyperP2, list(map(float,[F2[0][0], F2[1][0], F2[2][0]])), label="accuracy")
ax2.plot(hyperP2, list(map(float,[F2[0][1], F2[1][1], F2[2][1]])), label="loss")
# plt.xticks(Cx[1])
plt.xlabel("batch_size")
plt.legend()

ax3 = plt.subplot(223)
ax3.set_title("mo")
ax3.plot(hyperP3, list(map(float,[F3[0][0], F3[1][0], F3[2][0]])), label="accuracy")
ax3.plot(hyperP3, list(map(float,[F3[0][1], F3[1][1], F3[2][1]])), label="loss")
# plt.xticks(Cx[2])
plt.xlabel("momentum")
plt.legend()

ax4 = plt.subplot(224)
ax4.set_title("we")
ax4.plot(hyperP4, list(map(float,[F4[0][0], F4[1][0], F4[2][0]])), label="accuracy")
ax4.plot(hyperP4, list(map(float,[F4[0][1], F4[1][1], F4[2][1]])), label="loss")
# plt.xticks(Cx[3])
plt.xlabel("weight_decay")
plt.legend()

```

```python
f1 = open("logs/lrT.txt", "r", encoding="utf-8")
f2 = open("logs/bsT.txt", "r", encoding="utf-8")
f3 = open("logs/moT.txt", "r", encoding="utf-8")
f4 = open("logs/weT.txt", "r", encoding="utf-8")


F1 = f1.readlines()
F2 = f2.readlines()
F3 = f3.readlines()
F4 = f4.readlines()

f1.close()
f2.close()
f3.close()
f4.close()

F1 = [line.strip() for line in F1 if line.strip()!=""]
F2 = [line.strip() for line in F2 if line.strip()!=""]
F3 = [line.strip() for line in F3 if line.strip()!=""]
F4 = [line.strip() for line in F4 if line.strip()!=""]

print(F1)

y = [i for i in range(80)]

print(y)
plt.figure(figsize=(20, 13))
plt.figure(1)


x1 = F1[:80]
x1 = list(map(float, x1))
x2 = F1[80:160]
x2 = list(map(float, x2))
x3 = F1[160:]
x3 = list(map(float, x3))
ax1 = plt.subplot(221)
ax1.set_title("lr")
ax1.plot(y, x1, label="0.0001")
ax1.plot(y, x2, label="0.001")
ax1.plot(y, x3, label="0.01")
# plt.xticks(Cx[0])
# ax1.axes.yaxis.set_visible(False)
plt.grid(True)
plt.xlabel("lr")
plt.ylabel("loss")
plt.legend()

x4 = F2[:230]
x4 = list(map(float, x4))
x5 = F2[230:310]
x5 = list(map(float, x5))
x6 = F2[310:]
x6 = list(map(float, x6))
ax2 = plt.subplot(222)
ax2.set_title("batch size")
ax2.plot([i for i in range(230)], x4, label="32")
ax2.plot([i for i in range(80)], x5, label="64")
ax2.plot([i for i in range(30)], x6, label="128")
# ax2.axes.yaxis.set_visible(False)
plt.grid(True)
plt.xlabel("batch size")
plt.ylabel("loss")
plt.legend()

x7 = F3[:80]
x7 = list(map(float, x7))
x8 = F3[80:160]
x8 = list(map(float, x8))
x9 = F3[160:]
x9 = list(map(float, x9))
ax3 = plt.subplot(223)
ax3.set_title("momentum")
ax3.plot(y, x7, label="0.9")
ax3.plot(y, x8, label="0.99")
ax3.plot(y, x9, label="0.999")
# ax3.axes.yaxis.set_visible(False)
plt.grid(True)
plt.xlabel("momentum")
plt.ylabel("loss")
plt.legend()

x10 = F4[:80]
x10 = list(map(float, x10))
x11 = F4[80:160]
x11 = list(map(float, x11))
x12 = F4[160:]
x12 = list(map(float, x12))
ax4 = plt.subplot(224)
ax4.set_title("weight decay")
ax4.plot(y, x10, label="0.0001")
ax4.plot(y, x11, label="0.001")
ax4.plot(y, x12, label="0.01")
# ax4.axes.yaxis.set_visible(False)
plt.grid(True)
plt.xlabel("weight decay")
plt.ylabel("loss")
plt.legend()
plt.show()

```
